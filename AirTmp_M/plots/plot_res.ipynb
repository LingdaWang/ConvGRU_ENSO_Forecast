```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
PC = np.load('PC.npy')
RMSE = np.load('RMSE.npy')
```

```python
plt.figure()
l1 = plt.plot([i for i in range(1, 13)], PC,
              color='red', marker="o", markersize=10,
              linewidth=3.0, linestyle='-',
              label='ConvGRU')
plt.legend(prop={'size':15})
plt.xlabel('Lead (months)', fontdict={'size':15})
plt.ylabel('PC', fontdict={'size':15})
plt.grid('on')
# plt.show()
plt.tight_layout()
plt.savefig('Air_Temp_PC.pdf', bbox_inches='tight')
```

```python
plt.figure()
l1 = plt.plot([i for i in range(1, 13)], RMSE,
              color='blue', marker="o", markersize=10,
              linewidth=3.0, linestyle='-',
              label='ConvGRU')
plt.legend(prop={'size':15})
plt.xlabel('Lead (months)', fontdict={'size':15})
plt.ylabel('RMSE', fontdict={'size':15})
plt.grid('on')
# plt.show()
plt.savefig('Air_Temp_RMSE.pdf')
```

```python
ori_res = np.load('ori_res.npy')
pred_res = np.load('pred_res.npy')
```

```python
fig, axs = plt.subplots(4, 3)
i = 0
for row in range(4):
    for col in range(3):
        ax = axs[row, col]
        ax.axis('off')
        pcm = ax.imshow(ori_res[96, i, :, :])
        i += 1
fig.colorbar(pcm, ax=axs.ravel().tolist())
plt.savefig("sample_truth.png")
```

```python
fig, axs = plt.subplots(4, 3)
i = 0
for row in range(4):
    for col in range(3):
        ax = axs[row, col]
        ax.axis('off')
        pcm = ax.imshow(pred_res[96, i, :, :])
        i += 1
fig.colorbar(pcm, ax=axs.ravel().tolist())
plt.savefig("sample_pred.png")
```

```python
fig, axs = plt.subplots(4, 3)
i = 0
for row in range(4):
    for col in range(3):
        ax = axs[row, col]
        ax.axis('off')
        pcm = ax.imshow(pred_res[96, i, :, :] - ori_res[96, i, :, :])
        i += 1
fig.colorbar(pcm, ax=axs.ravel().tolist())
# plt.savefig("sample_pred.png")
plt.show()
```
