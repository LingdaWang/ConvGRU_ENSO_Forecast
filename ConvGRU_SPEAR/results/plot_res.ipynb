```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
P_corr_LR = np.load("res_LR/P_corr_LR.npy")
RMSE_LR = np.load("res_LR/RMSE_LR.npy")
P_corr_LIM = np.load("res_LIM/P_corr_LIM.npy")
RMSE_LIM = np.load("res_LIM/RMSE_LIM.npy")
P_corr_ConvGRU = np.load("res_ConvGRU/P_corr_ConvGRU.npy")
RMSE_ConvGRU = np.load('res_ConvGRU/RMSE_ConvGRU.npy')
P_corr_GRU = np.load("res_GRU/P_corr_GRU.npy")
RMSE_GRU = np.load('res_GRU/RMSE_GRU.npy')
```

```python
plt.figure()
l1 = plt.plot([i for i in range(1, 25)], P_corr_ConvGRU,
              color='blue', marker="o", markersize=6,
              linewidth=3.0, linestyle='-',
              label='ConvGRU')
l2 = plt.plot([i for i in range(1, 25)], P_corr_LIM,
              color='red', marker="d", markersize=6,
              linewidth=3.0, linestyle='-',
              label='LIM')
l3 = plt.plot([i for i in range(1, 25)], P_corr_GRU,
              color='green', marker="s", markersize=6,
              linewidth=3.0, linestyle='-',
              label='GRU')
l4 = plt.plot([i for i in range(1, 25)], P_corr_LR,
              color='black', marker="<", markersize=6,
              linewidth=3.0, linestyle='-',
              label='LR')
plt.legend(prop={'size':15})
plt.xlabel('Lead (months)', fontdict={'size':15})
plt.ylabel('PC', fontdict={'size':15})
plt.grid('on')
plt.savefig('SPEAR_PC_all.pdf')
# plt.show()
```
